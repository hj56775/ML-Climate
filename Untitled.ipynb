```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
data_train = pd.read_csv('DailyDelhiClimateTrain.csv',parse_dates=['date'],index_col='date')
data_test = pd.read_csv('DailyDelhiClimateTest.csv',parse_dates=['date'],index_col='date')
data_train.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(data_train['wind_speed'])
```

```python
#상관계수
plt.figure(figsize=(10,10))
sns.heatmap(data = data_train.corr(), annot=True, 
fmt = '.2f',square='True',linewidths=.5, cmap='Blues')
```

```python
#인과관계

from statsmodels.tsa.stattools import grangercausalitytests

variables = data_train.columns
maxlag=12
test = 'ssr_chi2test'

cause = pd.DataFrame(np.zeros((len(variables),len(variables))),columns=variables,index=variables)
for c in variables:
    for r in variables:
        x = grangercausalitytests(data_train[[r,c]],maxlag=maxlag,verbose=False)
        p_values = [round(x[i+1][0][test][1],5) for i in range(maxlag)]
        min_value = np.min(p_values)
        cause.loc[r,c] = min_value

cause

plt.figure(figsize=(10,10))
sns.heatmap(data = cause, annot=True, 
fmt = '.8f',square='True',linewidths=.5, cmap='Blues')
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
normalization_train_temp=sc.fit_transform(data_train.values[:,0].reshape([-1,1]))
normalization_test_temp=sc.fit_transform(data_test.values[:,0].reshape([-1,1]))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping,ModelCheckpoint

es= EarlyStopping(monitor='loss',mode='min',verbose=1,patience=5)

model1=Sequential()
model1.add(Dense(32, input_shape=(4,), activation='relu'))
model1.add(Dense(16, activation='relu'))
model1.add(Dense(8, activation='relu'))
model1.add(Dense(4, activation='relu'))
model1.add(Dense(1))

model1.summary()
```

```python
model1.compile(loss='mean_squared_error', optimizer='adam')
```

```python
#전날에 기온, 습도, 풍속, 기압을 이용해 다음날의 온도를 예측 
X_train=np.asarray(normalization_train[0:len(data_train)-2])#13년 1월 1일 부터 16년 12월 31일까지 4가지 변수
y_train=np.asarray(normalization_train[1:len(data_train)-1])[:,0]#13년 1월 2일 부터 17년 1월 1일까지 기온
```

```python
hist = model1.fit(X_train, y_train, batch_size=16, epochs=1000,callbacks=[es])
```

```python
X_test=np.asarray(normalization_test[0:len(data_test)-2]).astype('float32')#17년 1월 1일부터 17년 4월 23일까지의 4가지 변수
y_test=np.asarray(normalization_test[1:len(data_test)-1]).astype('float32')[:,0]#17년 1월 2일부터 17년 4월 24일까지의 기온
```

```python
trainScore=model1.evaluate(X_train, y_train, batch_size=16, verbose=0)
print('Train Score :', trainScore)

testScore=model1.evaluate(X_test, y_test, batch_size=1, verbose=0)
print('Test Score :', testScore)
```

```python
plt.plot(hist.history['loss'],c='r',label='loss')
plt.show()
```

```python
model1_pred=model1.predict(X_test,batch_size=16)
print(model1_pred[0:10])
```

```python
plt.figure(figsize = (15,5))
plt.plot(sc.inverse_transform(model1_pred),'b')
plt.plot(sc.inverse_transform(y_test.reshape([-1,1])),'y')
plt.show()
```
